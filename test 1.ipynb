```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
# Replace 'file_path' with the actual path to your CSV file
file_path = 'D:\jupyternotebooks\gdp.csv'
data = pd.read_csv(file_path)

# Filter data for Tanzania from 1990 to 2020
tanzania_data = data[data['Country'] == 'Tanzania'].iloc[:, 10:31]  # Assuming the years are in columns 10 to 30

# Reshape the data for plotting
tanzania_data = tanzania_data.T
tanzania_data.columns = ['GDP']  # Rename the column for better labeling

# Reset index to have years as a column
tanzania_data.reset_index(inplace=True)
tanzania_data.columns = ['Year', 'GDP']

# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='GDP', data=tanzania_data)
plt.title('Tanzania GDP Trend (1990-2020)')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load GDP data
gdp_data = pd.read_csv('gdp.csv')

# Load country to continent mapping data
country_continent_data = pd.read_csv('country_continent.csv')

# Merge GDP data with continent data based on country names
merged_data = pd.merge(gdp_data, country_continent_data, on='Country')

# Group data by continent and sum across all years
continent_gdp = merged_data.groupby('Continent').sum().drop(columns='Country')

# Transpose data for plotting
continent_gdp = continent_gdp.T.reset_index()

# Plotting using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='index', y='value', hue='Continent', data=pd.melt(continent_gdp, id_vars='index'))
plt.title('GDP Trend Across Continents')
plt.xlabel('Year')
plt.ylabel('GDP')
plt.legend(title='Continent')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load GDP data
gdp_data = pd.read_csv('gdp.csv')

# Load country to continent mapping data
country_continent_data = pd.read_csv('country_continent.csv')

# Merge GDP data with continent data based on country names
merged_data = pd.merge(gdp_data, country_continent_data, on='Country', how='left')

# Save the merged data with Continent column as a new CSV file
merged_data.to_csv('gdp_with_continent.csv', index=False)

print("New CSV 'gdp_with_continent.csv' created successfully.")
```
